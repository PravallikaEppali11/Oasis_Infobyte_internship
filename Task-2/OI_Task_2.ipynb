UNEMPLOYMENT ANALYSIS WITH PYTHON

Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the unemployment data into a Pandas DataFrame

```python
Unemployment_data = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
Unemployment_data.head()
```

Explore the data

```python
print(Unemployment_data.head())
print(Unemployment_data.info())
print(Unemployment_data.describe())
```

Visualization

```python
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
correlation_matrix = Unemployment_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

let’s see the unemployment rate according to different regions of India:

```python
Unemployment_data.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate","Estimated Employed",
               "Estimated Labour Participation Rate","Region",
               "longitude","latitude"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue="Region", data=Unemployment_data)
plt.show()
```

let's create a dashboard to analyze the unemployment rate of each Indian state by region using a sunburst plot, you can follow these steps:

```python
unemploment = Unemployment_data[["States", "Region", "Estimated Unemployment Rate"]]
figure = px.sunburst(unemploment, path=["Region", "States"],
                     values="Estimated Unemployment Rate",
                     width=700, height=700, color_continuous_scale="RdY1Gn",
                     title="Unemployment Rate in India")
figure.show()
```
