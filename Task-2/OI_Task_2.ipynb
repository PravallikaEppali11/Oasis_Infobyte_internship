UNEMPLOYMENT ANALYSIS WITH PYTHON

Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the unemployment data into a Pandas DataFrame

```python
Unemployment_data = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
Unemployment_data.head()
```

Explore the data

```python
print(Unemployment_data.head())
print(Unemployment_data.info())
print(Unemployment_data.describe())
```

Visualization

```python
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
correlation_matrix = Unemployment_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
